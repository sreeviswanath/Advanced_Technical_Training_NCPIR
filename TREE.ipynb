## Secret sharing(2327)

```python
class Solution:
    def peopleAwareOfSecret(self, n: int, delay: int, forget: int) -> int:
        MOD = 10**9 + 7
        # dp[i]: Number of people who learn the secret on day i (1-indexed)
        dp = [0] * (n + 1)
        dp[1] = 1  # Day 1: one person learns the secret

        # total_sharers: Count of people currently able to share the secret
        total_sharers = 0
        answer = 0

        for day in range(2, n + 1):
            # If day > delay: those who learned on day - delay start sharing today
            if day - delay >= 1:
                total_sharers = (total_sharers + dp[day - delay]) % MOD
            # If day > forget: those who learned on day - forget forget today
            if day - forget >= 1:
                total_sharers = (total_sharers - dp[day - forget] + MOD) % MOD

            dp[day] = total_sharers

        # Sum up people who still remember the secret on day n:
        for day in range(n - forget + 1, n + 1):
            if day >= 1:
                answer = (answer + dp[day]) % MOD

        return answer
obj=Solution()
obj.peopleAwareOfSecret(6,2,4)
```

## Count partitions(3578)

```python
from bisect import bisect_left, insort
from typing import List

class Solution:
    def countPartitions(self, nums: List[int], k: int) -> int:
        MOD = 10**9 + 7

        class SortedList:
            """Simple sorted list to maintain a sliding window."""
            def __init__(self):
                self.data = []

            def add(self, x: int):
                insort(self.data, x)

            def remove(self, x: int):
                idx = bisect_left(self.data, x)
                if idx < len(self.data) and self.data[idx] == x:
                    self.data.pop(idx)

            def get_min(self) -> int:
                return self.data[0] if self.data else float('inf')

            def get_max(self) -> int:
                return self.data[-1] if self.data else float('-inf')

        n = len(nums)
        dp = [0] * (n + 1)   # dp[i]: number of ways to partition first i elements
        prefix_sum = [0] * (n + 1)
        dp[0] = 1           # Base case: empty partition
        prefix_sum[0] = 1

        window = SortedList()
        left = 1  # 1-based index for left boundary of valid window

        for right in range(1, n + 1):
            window.add(nums[right - 1])

            # Shrink window until max-min â‰¤ k
            while window.get_max() - window.get_min() > k:
                window.remove(nums[left - 1])
                left += 1

            # Calculate dp[right]
            # Ways to partition ending at right = sum(dp[l - 1] ... dp[right - 1])
            sum_before_left = prefix_sum[left - 2] if left >= 2 else 0
            dp[right] = (prefix_sum[right - 1] - sum_before_left + MOD) % MOD
            prefix_sum[right] = (prefix_sum[right - 1] + dp[right]) % MOD

        return dp[n]

obj=Solution()
obj.countPartitions([9,4,1,3,7],4)
```

# K TH SMALLEST ON LEFT AND RIGHT SUBTREE

```python
from collections import deque

class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class Tree:
    def __init__(self):
        self.root = None

    def append(self, data):
        newnode = Node(data)
        if self.root is None:
            self.root = newnode
        else:
            temp = self.root
            while True:
                if data < temp.data:
                    if temp.left is not None:
                        temp = temp.left
                    else:
                        temp.left = newnode
                        break
                else:
                    if temp.right is not None:
                        temp = temp.right
                    else:
                        temp.right = newnode
                        break

    def treeheight(self, root):
        if root is None:
            return 0
        lheight = self.treeheight(root.left)
        rheight = self.treeheight(root.right)
        return max(lheight, rheight) + 1

    def kth_smallest(self, root, k):

        stack = []
        current = root
        count = 0

        while True:
            while current:
                stack.append(current)
                current = current.left

            if not stack:
                return None  # k is larger than the size of the subtree

            current = stack.pop()
            count += 1
            if count == k:
                return current.data

            current = current.right

    def kth_smallest_subtrees(self, k):
        if self.root is None:
            print("Tree is empty")
            return

        left_result = self.kth_smallest(self.root.left, k) if self.root.left else None
        right_result = self.kth_smallest(self.root.right, k) if self.root.right else None

        print(f"{k}-th smallest in LEFT subtree: {left_result}")
        print(f"{k}-th smallest in RIGHT subtree: {right_result}")



obj = Tree()
while True:
    data = int(input("Enter value (-1 to stop): "))
    if data <= 0:
        break
    obj.append(data)

h = obj.treeheight(obj.root)
print(f"HEIGHT = {h}")

k = int(input("Enter k: "))
obj.kth_smallest_subtrees(k)
```

## Distance between nodes

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class Tree:
    def __init__(self):
        self.root = None

    def append(self, data):
        newnode = Node(data)
        if self.root is None:
            self.root = newnode
        else:
            temp = self.root
            while True:
                if data < temp.data:
                    if temp.left is not None:
                        temp = temp.left
                    else:
                        temp.left = newnode
                        break
                else:
                    if temp.right is not None:
                        temp = temp.right
                    else:
                        temp.right = newnode
                        break

    def find_LCA(self, root, n1, n2):

        while root:
            if n1 < root.data and n2 < root.data:
                root = root.left
            elif n1 > root.data and n2 > root.data:
                root = root.right
            else:
                return root
        return None

    def distance_from_root(self, root, val):

        distance = 0
        while root and root.data != val:
            if val < root.data:
                root = root.left
            else:
                root = root.right
            distance += 1
        return distance

    def distance_between_nodes(self, n1, n2):

        lca = self.find_LCA(self.root, n1, n2)
        if not lca:
            return -1
        d1 = self.distance_from_root(lca, n1)
        d2 = self.distance_from_root(lca, n2)
        return d1 + d2



obj = Tree()
while True:
    data = int(input("Enter value (-1 to stop): "))
    if data <= 0:
        break
    obj.append(data)

n1 = int(input("Enter first node: "))
n2 = int(input("Enter second node: "))

distance = obj.distance_between_nodes(n1, n2)
print(f"Distance between {n1} and {n2} is {distance}")
```

## Mirror of BST

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class Tree:
    def __init__(self):
        self.root = None

    def append(self, data):
        newnode = Node(data)
        if self.root is None:
            self.root = newnode
        else:
            temp = self.root
            while True:
                if data < temp.data:
                    if temp.left is not None:
                        temp = temp.left
                    else:
                        temp.left = newnode
                        break
                else:
                    if temp.right is not None:
                        temp = temp.right
                    else:
                        temp.right = newnode
                        break

    def mirror(self, node):
        if node is None:
            return

        node.left, node.right = node.right, node.left

        self.mirror(node.left)
        self.mirror(node.right)

    def inorder(self, node):
        if node:
            self.inorder(node.left)
            print(node.data, end=" ")
            self.inorder(node.right)


tree = Tree()
while True:
    data = int(input("Enter node value (-1 to stop): "))
    if data == -1:
        break
    tree.append(data)

print("Original BST (inorder):")
tree.inorder(tree.root)

tree.mirror(tree.root)

print("\nMirrored BST (inorder):")
tree.inorder(tree.root)
```

# VERTICAL ORDER TRAVERSAL

```python
from collections import deque
from collections import defaultdict
class Node:
  def __init__(self, data):
    self.data = data
    self.left = None
    self.right = None
class Tree:
  def __init__(self):
    self.root = None
  def append(self,data):
    newnode=Node(data)
    if self.root is None:
      self.root=newnode
    else:
      temp=self.root
      while True:
        if data<temp.data:
          if temp.left!=None:
            temp=temp.left
          else:
            temp.left=newnode
            break
        else:
          if temp.right!=None:
            temp=temp.right
          else:
            temp.right=newnode
            break


  def verticalTraversal(self):
    que=deque()
    que.append([self.root,0])
    result=defaultdict(list)
    while que:
      currnode,Hd=que.popleft()
      result[Hd].append(currnode.data)
      if currnode.left:
        que.append([currnode.left,Hd-1])
      if currnode.right:
        que.append([currnode.right,Hd+1])
    for Hd in sorted(result.keys()):
      print(f"{Hd}->{result[Hd]}")


obj=Tree()
while True:
  data=int(input())
  if data<=0:
    break
  obj.append(data)
obj.verticalTraversal()
```
