# SINGLY LINKED LIST


```python
class Node:
  def __init__(self, d):
    self.data = d
    self.add = None

class LinkedList:
  def __init__(self):
    self.head = None
  def append(self,data):
    new_node = Node(data)
    if self.head==None:
      self.head = new_node
    else:
      temp=self.head
      while temp.add!=None:
        temp=temp.add
      temp.add=new_node
  def display(self):
    temp=self.head
    while temp!=None:
      print(temp.data,end=" ")
      temp=temp.add

list=LinkedList()
while(True):
  try:
    user_input = int(input("Enter input (or a negative num to stop): "))
    if user_input < 0:
      break
    list.append(user_input)
  except ValueError:
    print("Invalid input. Please enter an integer.")

list.display()


```

# DOUBLY LINKED LIST

```python
class Node:
  def __init__(self, d):
    self.data = d
    self.add = None
    self.right=None
    self.left=None

class LinkedList:
  def __init__(self):
    self.head = None
    self.tail=None
  def append(self,data):
    new_node = Node(data)
    if self.head==None:
      self.head = new_node
    else:
      temp=self.head
      while temp.right!=None:
        temp=temp.right
      temp.right=new_node
      new_node.left=temp
      self.tail=new_node
  def forwarddisplay(self):
    temp=self.head
    print("FORWARD ITERATE:")
    while temp!=None:
      print(temp.data,end=" ")
      temp=temp.right
    print()
  def backwarddisplay(self):
    print("BACKWARD ITERATE:")
    temp=self.tail
    while temp!=None:
      print(temp.data,end=" ")
      temp=temp.left

list=LinkedList()
while(True):
  try:
    user_input = int(input("Enter input (or a negative num to stop):"))
    if user_input < 0:
      break
    list.append(user_input)
  except ValueError:
    print("Invalid input. Please enter an integer.")

list.forwarddisplay()
list.backwarddisplay()
```
