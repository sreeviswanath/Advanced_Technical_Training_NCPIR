## circle game (1823)

```python
from collections import deque
class Solution(object):
    def findTheWinner(self, n, k):
        queue=deque()
        for i in range(1,n+1):
            queue.append(i)
        while len(queue)!=1:
            for i in range(1,k+1):
                d=queue.popleft()
                if i!=k:
                    queue.append(d)

        return queue[-1]
n=int(input("n="))
k=int(input("k="))

result= Solution()
winner = result.findTheWinner(n, k)
print("Winner=",end=" ")
print(winner)
```

# Rotting oranges(994)

```python
from collections import deque
class Solution(object):
    def orangesRotting(self, grid):
        row=len(grid)
        col=len(grid[0])
        min=0
        fresh_org=0
        que=deque()
        for i in range(row):
            for j in range(col):
                if grid[i][j]==1:
                    fresh_org+=1
                elif grid[i][j]==2:
                    que.append((i,j))
        dir=[[0,-1],[0,1],[-1,0],[1,0]]
        while que and fresh_org>0:
            size=len(que)
            for i in range(size):
                x,y=que.popleft()
                for dx,dy in dir:
                    nx=x+dx
                    ny=y+dy

                    if nx>=0 and ny>=0 and nx<row and ny<col and grid[nx][ny]==1:
                        fresh_org-=1
                        grid[nx][ny]=2
                        que.append((nx,ny))
            min+=1
        if fresh_org==0:
            return min
        else:
            return -1

obj=Solution()
grid=[[2,1,1],[1,1,0],[0,1,1]]
result=obj.orangesRotting(grid)
print(result)
```

## Robotic vaccum cleaner route

```python
from collections import deque

def count_reachable_tiles(grid):
    rows = len(grid)
    cols = len(grid[0])
    visited = [[False] * cols for _ in range(rows)]
    q = deque()

    start = None
    for i in range(rows):
        for j in range(cols):
            if grid[i][j] == 2:
                start = (i, j)
                break
        if start:
            break

    if not start:
        return 0

    q.append(start)
    visited[start[0]][start[1]] = True
    reachable_count = 0

    directions = [(0, 1), (1, 0), (0, -1), (-1, 0)]

    while q:
        x, y = q.popleft()
        reachable_count += 1

        for dx, dy in directions:
            nx, ny = x + dx, y + dy
            if 0 <= nx < rows and 0 <= ny < cols:
                if not visited[nx][ny] and grid[nx][ny] != 1:
                    visited[nx][ny] = True
                    q.append((nx, ny))

    return reachable_count


floor_plan = [
    [2,0,1],
    [0,0,0],
    [1,0,1]
]

result=count_reachable_tiles(floor_plan)
print(f"Number of reachable tiles = {result}")
```
