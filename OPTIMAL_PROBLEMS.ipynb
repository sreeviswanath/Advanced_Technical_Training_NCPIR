# GREEDY ALGORITHM

# COINS AND CHANGE PROBLEM

# TOTAL NO.OF WAYS

```python
coins=[1,2,3]
change=5
N=len(coins)
dp=[[0]*(change+1) for _ in range(N+1)]
for i in range(N+1):
  dp[i][0]=1
for i in range(1,N+1):
  for j in range(1,change+1):

    if j<coins[i-1]:
      dp[i][j]=dp[i-1][j]

    else:
      dp[i][j]=dp[i-1][j]+dp[i][j-coins[i-1]]

for i in range(N+1):
  for j in range(change+1):
    print(dp[i][j],end=" ")
  print()
print()
print("Total no.of ways shopkeeper can give change =",dp[N][change])
```

# DYNAMIC PROGRAMMING

# MINIMUM NUMBER OF COINS

```python
coins=[1,5,6,9]
change=11
N=len(coins)
dp=[[float('inf')]*(change+1) for _ in range(N+1)]
for i in range(N+1):
  dp[i][0]=0
for i in range(1,N+1):
  for j in range(1,change+1):

    if j<coins[i-1]:
      dp[i][j]=dp[i-1][j]

    else:
      dp[i][j]=min(dp[i-1][j],1+dp[i][j-coins[i-1]])

for i in range(1,N+1):
  for j in range(1,change+1):
    print(dp[i ][j ],end="   ")
  print()
print()
print("Minimum coins shopkeeper can give change =",dp[N][change])
```

# 0-1 KNAPSACK

```python
weight=[3,4,5,6]
price=[2,3,4,1]
capacity=8
N=len(weight)
dp=[[0]*(capacity+1) for _ in range(N+1)]
for i in range(N+1):
  dp[i][0]=0
for i in range(1,N+1):
  for j in range(capacity+1):

    if j<weight[i-1]:
      dp[i][j]=dp[i-1][j]

    else:
      dp[i][j]=max(dp[i-1][j],dp[i-1][j-weight[i-1]]+price[i-1])

for i in range(N+1):
  for j in range(capacity+1):
    print(dp[i][j],end=" ")
  print()
print()
print("MAXIMUM PROFIT =",dp[N][capacity])
```
